```python
def df_explore(df):
    print('Shape: ', df.shape)
    
    t=['Columns','Type','Non-Null','Null','Unique','Example']
    
    for c in df.columns:
        t.add_row([c,
                  df[c].dtype,
                  len(df[c])-np.sum(df[c].isna()),
                  np.sum(df[c].isna()),
                   np.count_nonzero(df[c],unique()),
                   df[~df[c].isnull()][c].iloc[0],
                   ])
        print(t)
        print()
        return
    
    df_explore(df)
```

```python
def column_exploration(df,column_name,sort_index=False):
    #Analysis of column values and plot
    print(f'COLUMN: {column_name}')

    # Quantity and Percentage of orders
    print('\n\na. Quantity and Percentage \n')
    abs_values =df[column_name].value_counts()
    prc_values =(df[column_name].value_counts(normalize=True)*100).apply(lambda x:round(x,1))
    df_values = pd.merge(abs_values, prc_values, left_index=True, right_index=True, suffixes=('_abs','_prc'))
    #display(df_values)

    # Plot of orders
    #print('\n\nb. Plot of quantity\n')
    
    if sort_index is True:
        df_values = df_values.sort_index(axis = 0)
    
    # Plot values, title and axis names
    ax = df_values[column_name+'_abs'].plot(kind='bar', figsize=(15,7))
    ax.set_title(f"Cantidad by {column_name}")
    ax.set_xlabel(column_name)
    ax.set_ylabel('Cantidad')

    #Labels over bars
    rects = ax.patches
    labels = [f'{p}%' for p in df_values[column_name+'_prc'].to_list()]
    for rect, label in zip(rects, labels):
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width() / 2, height + 5, label,
                ha='center', va='bottom')
    
    return
```


​Merge con apertura < llamada

```python
import datetime
print((datetime.datetime.now()))
```

```python
merge['FECHA']=pd.to_datetime(merge['FECHA'], format="%Y/%m/%d")
```

```python
merge['Fecha de creación']=pd.to_datetime(merge['Fecha de creación'], format="%d/%m/%Y %H:%M")
```

```python
merge['dias']= (merge['Fecha de creación'] - merge['FECHA'])
```

```python
filtro0= merge['dias'] > '0 days'
filtro10= merge['dias'] <= '10 days'
merge10dias=merge[filtro0 & filtro10]
merge10dias.head()
```

```python
merge10dias.shape
```

De los que llamaron y recibieron email, solo se toman los que hicieron llamada en los siguientes 10 días luego de recibir el email. En ese caso, solo hay 91 clientes. Es baja la tasa de llamadas respecto a los emails enviados, pero es correcto, ya que es una campaña de bienvenida a la empresa y no debería generar dudas en los clientes.

```python
plt.figure(figsize=(20,20))
vars_to_plot = ['Tipo Contacto','Motivo','Motivo Generico','Retenible']
for i, var in enumerate(vars_to_plot):
    plt.subplot(2, 2, i+1)
    value_counts=merge10dias[var].value_counts()
    ax=sns.barplot(x=value_counts.index, y=value_counts.values)
    title_string='countplot'+ var
    plt.ylabel("count")
    plt.title(title_string)
    ax.set_xticklabels(ax.get_xticklabels(),rotation=30)
```

Los clientes llamaron principalmente para conocer más sobre la empreza y conocer la cobertura de su producto.  Se debe poner atención en los grupos que llamaron por "venta" y "bajas". Los que llamaron por baja es porque ya no quieren el producto.  Los que llamaron por venta, hay que evaluar que pasó con su venta y el motivo de insatisfacción que puedan tener.

```python
crosstab=pd.crosstab(index=[merge10dias['Tipo Contacto']], columns=merge10dias['CAMPANIA'])
crosstab
cm=sns.light_palette("blue", as_cmap=True)
crosstab.style.background_gradient(cmap=cm)
```

```python
crosstab2=pd.crosstab(index=[merge10dias['Tipo Contacto'],merge10dias['Motivo Generico']], columns=merge10dias['CAMPANIA'])
crosstab2
cm=sns.light_palette("blue", as_cmap=True)
crosstab2.style.background_gradient(cmap=cm)
```

```python
crosstab2=pd.crosstab(index=[merge10dias['Tipo Contacto'],merge10dias['Motivo Generico'], merge10dias['Motivo']], columns=merge10dias['CAMPANIA'])
crosstab2
cm=sns.light_palette("blue", as_cmap=True)
crosstab2.style.background_gradient(cmap=cm)
```

Los que llamaron por venta, dicen que desconocen la adhesion al producto o tienen disconformidad con la gestión de venta. Averiguar más detalle.

```python
crosstab3=pd.crosstab(index=[merge10dias['Motivo Generico'],merge10dias['Observaciones']], columns=merge10dias['CAMPANIA'])
crosstab3
cm=sns.light_palette("blue", as_cmap=True)
crosstab3.style.background_gradient(cmap=cm)
```

Los clientes que llamaron por problemas con la venta, todos se dieron de baja. Aunque su llamada no fue directamente para pedir la baja, sino para entender por qué los habían contactado, terminaron por decidir solicitar la baja.
